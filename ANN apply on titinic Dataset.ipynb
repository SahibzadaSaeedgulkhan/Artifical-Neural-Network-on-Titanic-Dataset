# ANN

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(units = 10, input_shape = (3,),activation = 'relu'))    # Input layer
          
model.add(Dense(units = 6, activation = 'relu'))    # hidden layer
          
model.add(Dense(1, activation = 'sigmoid')) # Output layer
          
          
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])  # compiling model
          
model.summary()
```

```python
model.fit(xtrain,ytrain,validation_split=0.1,epochs=20)
```

```python
np.round(model.predict(xtest))
```

```python
np.round(ytest)
```

```python
model.evaluate(xtest,ytest)
```
